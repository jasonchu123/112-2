```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('adult.csv')
salary_data = df['income']
plt.figure(figsize=(10, 6))
sns.histplot(salary_data, bins=5, kde=True, color='skyblue')
plt.title('histogram')
plt.xlabel('income')
plt.ylabel('Age')
plt.show()
```
